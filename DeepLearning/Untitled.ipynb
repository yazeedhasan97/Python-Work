```python
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
def solution_model():
    xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
    ys = np.array([-7.0, -3.0, 1.0, 5.0, 9.0, 13.0], dtype=float)

    model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, activation='relu', input_shape=[1, ]),
        # tf.keras.layers.BatchNormalization(),
        # tf.keras.layers.Dense(1, activation='relu', ),
        # tf.keras.layers.BatchNormalization(),
        # tf.keras.layers.Dense(1, activation='relu', ),
        # tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(1),
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Nadam(learning_rate=1e-3),
        # optimizer=tf.keras.optimizers.Adam(),
        loss='mae',
        metrics=['mae', 'mse']
    )

    MONITOR = 'loss'
    BATCH_SIZE = 3
    EPOCHS = 500

    callbacks = [tf.keras.callbacks.ReduceLROnPlateau(monitor=MONITOR, factor=0.1, patience=10, min_lr=1e-5)]
    callbacks += [tf.keras.callbacks.EarlyStopping(monitor=MONITOR, patience=15, restore_best_weights=True)]

    history = model.fit(
        xs,
        ys,
        # validation_split=0.1,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=callbacks,
    )
    

    return model

```
