Mission Two

On the drive, you will find the file (short_movies.csv) this file contains some movies reviews which
were extracted from IMDb users posts using their APIs. From that file find the following:

1 - The html with biggest number of reviews.

```python
review_count = {}
with open('short_movies.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        if i['html'] in review_count:
            review_count[i['html']] += 1
        else:
            review_count[i['html']] = 1

def by_value(item):
    return item[-1]

print('The html with biggest number of reviews is:', max(review_count, key=by_value))
```

2 - A list represent the number of occurrences of the words good, fun(ny), and great for each movie.

    (those words considered as a good review of the movies)
    
3 - The html with the biggest number of occurrences of the words good, fun(ny), and great in its reviews.

```python
good_review = []
html_good_review_count = {}
with open('short_movies.csv') as file:
    csv_reader = csv.DictReader(file)
    for i in csv_reader:
        temp = re.findall(r'[Gg]ood|[Ff]unn?y?|[Gg]reat', i['text'])
        l = len(temp)
        good_review.append(l)
        
        if i['html'] in html_good_review_count:
            html_good_review_count[i['html']] += l
        else:
            html_good_review_count[i['html']] = l
    print(good_review)
    
    
    
    
print('The html with the biggest good reviews is:', max(html_good_review_count, key=by_value))
```

4 - A list represent the number of occurrences of the words bad, sad, and suck(s) for each movie..

    (those words considered as a bad review of the movies)
    
5 - The html with the biggest number of occurrences of the words bad, sad, and suck(s) in its reviews.

```python
bad_review = []
html_bad_review_count = {}
with open('short_movies.csv') as file:
    csv_reader = csv.DictReader(file)
    for i in csv_reader:
        temp = re.findall(r'[Bb]ad|[Ss]ad|[Ss]ucks?', i['text'])
        l = len(temp)
        bad_review.append(l)
        
        if i['html'] in html_bad_review_count:
            html_bad_review_count[i['html']] += l
        else:
            html_bad_review_count[i['html']] = l
    print(good_review)
    
    
    
    
print('The html with the biggest bad reviews is:', max(html_bad_review_count, key=by_value))
```

6 - A list represents the ratio between good reviews and bad reviews for each html.

```python
in_total_ratio = []
for total, good, bad in zip(review_count.values(), html_good_review_count.values(), html_bad_review_count.values()):
    in_total_ratio.append([total, good/total, bad/total])
in_total_ratio
```

7 - How many time a full number appeared in the text column of this data set.

```python
numbers_count = 0
with open('short_movies.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        st = i['text']
        li = re.findall(r"[\W]+(\d+)", st)
        numbers_count += len(li)
print('A full number appeared:', numbers_count, 'times')
```

8 - How many time a question mark appeared in the text column of this data set.

```python
questions_count = 0
with open('short_movies.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        li = re.findall(r"\?", i['text'])
        questions_count += len(li)
print('A full question mark appeared:', questions_count, 'times')
```

9 - List off all questions appeared in the text column of this data set.

```python
questions_count = 0
with open('short_movies.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        st = i['text']
        li = re.findall(r"[\W\.]+\s([\w\W]+?\?)\W{,3}", st)
        questions_count += len(li)
print('A full question appeared:', questions_count, 'times')
```

Mission Three

On the drive, you will find the file (short_tweets.csv) this file contains some tweets which were extracted from twitter using their APIs. From that file find the following:

1 - A list contains all the users mentions (the user mention look like @username).

```python
user_minsions = []
with open('short_tweets.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        st = i['text']
        li = re.findall(r"@[\w\W]+?\s", st)
        user_minsions.extend(li)
print(user_minsions)
```

2 - A list of all users contains a numbers in their names.

```python
users_with_numbers = []
with open('short_tweets.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        st = i['user']
        if re.findall(r"\d", st):
            users_with_numbers.append(st)
            
print(users_with_numbers)
```

3 - A list of the most frequent 50 words in the file (Uppercase or Lowercase) with the number of their occurrences.

```python
words_counter = {}
with open('short_tweets.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        st = re.findall(r'\s?([\w\W]+?)\s',i['text'])
        for word in st:
            if word in words_counter:
                words_counter[word] += 1
            else:
                words_counter[word] = 1
print('The most frequant 50 words are: ')
def by_value(x):
    return x[-1]

sor = list(words_counter.items())
sor.sort(key=by_value, reverse=True)
sor[:50]
```

4 - Which occurs more Love or Hate?!, Which occurs more Good or Bad?!

```python
love_count = 0
good_count = 0
with open('short_tweets.csv') as file:
    csv_reader = csv.DictReader(file)
    headers = csv_reader.fieldnames
    for i in csv_reader:
        st = re.findall(r'[Gg]ood',i['text'])
        good_count += len(st)
        st = re.findall(r'[Bb]ad',i['text'])
        good_count -= len(st)
        st = re.findall(r'[Ll]ove',i['text'])
        love_count += len(st)
        st = re.findall(r'[Hh]ate',i['text'])
        love_count -= len(st)
        
print('Love' if love_count > 0 else 'Hate', 'occurs more')     
print('Good' if good_count > 0 else 'Bad', 'occurs more')
```
