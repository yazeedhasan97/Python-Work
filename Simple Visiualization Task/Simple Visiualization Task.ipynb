```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,12)
%matplotlib inline

import seaborn as sns
sns.set()

import os
import glob
```

```python
path = os.path.abspath(os.getcwd())
os.chdir(path)
result = glob.glob('*.{}'.format('csv'))
print(result)
```

```python
for file in result:
    df = pd.read_csv(file, index_col='tbl_dt')
    df.head()

    plt.figure(figsize=(20,10))
    plt.plot(df)
    plt.yscale('log')
    plt.xticks(rotation=15)
    plt.legend(df.columns,loc='best')
    plt.gca().get_xaxis().get_major_formatter().set_useOffset(False)
    plt.savefig(file.replace('csv', 'png'), bbox_inches='tight')
    plt.show()

```
